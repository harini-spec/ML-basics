```python
# Example 2
# perfect square
# Anu is asked to find the square root of perfect square numbers from the given list. How can she find it using numpy?
# nums = [1, 16, 20, 25, 4]
```

```python
import numpy as np
nums = [1, 16, 18, 25, 4]
numbers = np.sqrt(nums)
#print(numbers)
```

```python
print("Square roots of perfect squares:")
for num in numbers:
    if num.is_integer():
        print(int(num))
```

```python
# Example 3
# Replace
# Reema is asked to replace all the odd numbers present in the range 1 to 10 with -1. Achieve the same using numpy
```

```python
import numpy as np
arr = np.arange(1, 11)
out = np.where(arr % 2 == 1, -1, arr)
print(arr)
print(out)
```

```python
# Example 6
# Arun wants to create an array which contains elements from array 1 and array 2 but none of the elements should appear more than once
# array 1: [2, 4, 6, 1, 3]
# array 2: [4, 8, 9, 1, 6]
```

```python
# Example 8
# Anand is asked to extract all the odd numbers from a random list generated using numpy. Help him achieve the same
```

```python
import numpy as np
array = np.random.randint(20, size=12)
print(array)
```

```python
cond = np.mod(array, 2)==1
print(cond)
```

```python
odd = np.extract(cond, array)
print(odd)
```

```python
# Example 9
# Arav is said that all the scores above 5 are considered as Hit and scores less than or equal to 5 are considered as Miss. 
# Help him categorize the scores as Hit or Miss
# list provided : [1,5,6,8,1,7,3,6,9]
```

```python
y = np.array([1,5,6,8,1,7,3,6,9])
final = np.where(y>5, "Hit", "Miss")
```

```python
for i in range (0,9):
    print(y[i] , ":" , final[i])
```
